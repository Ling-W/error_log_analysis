```python
import pandas as pd
```

```python
in_file_path = 'C:/Users/liwu/XstoreLogProject/data'
out_file_path = 'C:/Users/liwu/XstoreLogProject/data'
```

```python
class Process:
    def __init__(self,in_fp, out_fp):
        self.in_file_path = in_fp
        self.out_file_path = out_fp
    def process_timevec(self,in_filename,out_filename):
        df = pd.read_csv(self.in_file_path + '/' + in_filename)
        
        def timevec(x):
            vec = [0]*(24*7)
            for row in x.itertuples():
                pos = int(row.hour) + int(row.weekday)*24
                vec[pos] += 1
            return vec
        
        new_df = df.groupby('oracle_error_code').apply(timevec).reset_index()
        new_df = pd.DataFrame(new_df)
        new_df.rename(columns={ new_df.columns[1]: "timevec"}, inplace = True)
        print(new_df)
        new_df.to_csv( out_file_path + '/' + out_filename,index = False)
        
    def normalize(self,df):
        pd.processing.normalize(df.timevec,norm = 'l2')
        
    
        
```

```python
  P1 = Process(in_file_path, out_file_path)
P1.process_timevec('errorLog_v1.csv_clean_simplify_info.csv','errorLog_v1_timevec.csv')
  
```
